Importing the dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data collection and pre-processing

```python
# loading the data from the csv file to a pandas DataFrame
movies_data = pd.read_csv('/content/movies.csv')
```

```python
# print first 5 rows of our Data Frame
movies_data.head(n=5)
```

```python
# find number of rows and columns in the DataFrame
movies_data.shape
```

```python
# selecting relevant columns for recommendation
selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
# replace null values with null string
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combine the 5 selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
print(combined_features)
```

```python
#converting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# cosine similarity
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the movie name from the user
movie_name = input('Enter your favorite movie:')
```

```python
# create a list with all the movie names given in the dataset
# every title

list_movies = movies_data['title'].tolist()
print(list_movies)
```

```python
# Need to check if the name given is in the list
# Finding the close match for the name given by the user

find_close_match = difflib.get_close_matches(movie_name, list_movies)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# find index of movie with title given
index_of_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
#getting a list of similar movies
#looping the similarity score of each movie
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sort movies based on similarity scores
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

```python
# print the name of the similar movies based on the index
print('Movies Suggested for You: \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]
  if (i<30):
    print(i, '.', title_from_index)
    i+=1
```

Movie Recommendation System

```python
movie_name = input('Enter your favorite movie:')

list_movies = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_movies)

close_match = find_close_match[0]

index_of_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)

print('Movies Suggested for You: \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]
  if (i<30):
    print(i, '.', title_from_index)
    i+=1
```
